```python
import gym
from stable_baselines3 import DDPG
from stable_baselines3.common.vec_env import VecFrameStack
from stable_baselines3.common.evaluation import evaluate_policy
import os
```

```python
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.env_util import make_vec_env
```

```python
from gym.envs import box2d
```

```python
from stable_baselines3.common.noise import NormalActionNoise, OrnsteinUhlenbeckActionNoise
import numpy as np
```

```python
environment_name = "BipedalWalker-v3"
```

```python
env = gym.make(environment_name, hardcore=True)
```

```python
env.action_space.sample()
```

```python
vec_env = make_vec_env(environment_name, n_envs=4)
```

```python
log_path = os.path.join('Training', 'Logs')
```

```python
# The noise objects for DDPG
n_actions = env.action_space.shape[-1]
action_noise = NormalActionNoise(mean=np.zeros(n_actions), sigma=0.1 * np.ones(n_actions))
```

```python
model = DDPG("MlpPolicy", env, verbose=1, tensorboard_log=log_path, action_noise=action_noise)
```

```python
model.learn(total_timesteps=100000)
```

```python
ppo_path = os.path.join('Training', 'Saved Models', 'DDPG_Walker_model_100k')
```

```python
model.save(ppo_path)
```
